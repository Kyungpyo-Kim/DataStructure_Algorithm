## 4874. [파이썬 S/W 문제해결 기본] 5일차 - Forth

※ SW Expert 아카데미의 문제를 무단 복제하는 것을 금지합니다.


Forth라는 컴퓨터 언어는 스택 연산을 기반으로 하고 있어 후위 표기법을 사용한다. 예를 들어 3+4는 다음과 같이 표기한다.
 

3 4 + .
 

Forth에서는 동작은 다음과 같다.
 

숫자는 스택에 넣는다.

연산자를 만나면 스택의 숫자 두 개를 꺼내 더하고 결과를 다시 스택에 넣는다.

‘.’은 스택에서 숫자를 꺼내 출력한다.

 

Forth 코드의 연산 결과를 출력하는 프로그램을 만드시오. 만약 형식이 잘못되어 연산이 불가능한 경우 ‘error’를 출력한다.
 

다음은 Forth 연산의 예이다.
 

코드

출력

4 2 / .

2

4 3 - .

1

 

 

[입력]
 

첫 줄에 테스트 케이스 개수 T가 주어진다.  1≤T≤50
 

다음 줄부터 테스트 케이스의 별로 정수와 연산자가 256자 이내의 연산코드가 주어진다. 피연산자와 연산자는 여백으로 구분되어 있으며, 코드는 ‘.’로 끝난다.

나눗셈의 경우 항상 나누어 떨어진다.

 

[출력]
 

#과 1번부터인 테스트케이스 번호, 빈칸에 이어 계산결과를 정수로 출력하거나 또는 ‘error’를 출력한다.

```
입력:
3
10 2 + 3 4 + * .
5 3 * + .
1 5 8 10 3 4 + + 3 + * 2 + + + .

출력:
#1 84
#2 error
#3 168
```

````python
# Make input file
f = open("input.txt", "w")
f.write("3\n")
f.write("10 2 + 3 4 + * .\n")
f.write("5 3 * + .\n")
f.write("1 5 8 10 3 4 + + 3 + * 2 + + + .\n")
f.close()
````

````python
# for Jupyter Notebook
###
f = open("input.txt", "r")
input = f.readline
###

# Ref: https://github.com/pkch93/Algorithm/blob/master/SW_Expert_Academy/level2/forth.py
operators = {
    '+': lambda x, y: x + y,
    '-': lambda x, y: x - y,
    '*': lambda x, y: x * y,
    '/': lambda x, y: x // y,
}

T = int(input())

for test_case in range(1, T + 1):

    forthCode = list( input().split() )
    
    opers = operators.keys()
    
    forthStack = []
    numStack = []
    operStack = []
        
    for fc in forthCode:
        if fc in opers:
            if len(forthStack) < 2:
                forthStack = []
                break
            else:
                s = forthStack.pop()
                f = forthStack.pop()
                forthStack.append( operators[fc](f, s) )
                
        elif fc=='.':
            break
        else:
            forthStack.append(int(fc))
    
    result = None
    if not len(forthStack)==1:
        result = "error"
    else:
        result = forthStack[0]
        
    print( "#"+str(test_case), result)
````

## 4875. [파이썬 S/W 문제해결 기본] 5일차 - 미로

※ SW Expert 아카데미의 문제를 무단 복제하는 것을 금지합니다.


NxN 크기의 미로에서 출발지에서 목적지에 도착하는 경로가 존재하는지 확인하는 프로그램을 작성하시오. 도착할 수 있으면 1, 아니면 0을 출력한다.

주어진 미로 밖으로는 나갈 수 없다.
 

다음은 5x5 미로의 예이다.
 

13101

10101

10101

10101

10021

 

마지막 줄의 2에서 출발해서 0인 통로를 따라 이동하면 맨 윗줄의 3에 도착할 수 있는지 확인하면 된다.

 
 

[입력]
 

첫 줄에 테스트 케이스 개수 T가 주어진다.  1<=T<=50
 

다음 줄부터 테스트 케이스의 별로 미로의 크기 N과 N개의 줄에 걸쳐 미로의 통로와 벽에 대한 정보가 주어진다. 0은 통로, 1은 벽, 2는 출발, 3은 도착이다. 5<=N<=100

 

[출력]
 

각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, 계산결과를 정수로 출력하거나 또는 ‘error’를 출력한다.

```
입력:

5

10101
10101
10021
5
10031
10111
10101
10101
12001
5
00013

출력:
#1 1

```

````python
# Make input file
f = open("input.txt", "w")
f.write("3\n")
f.write("5\n")
f.write("13101\n")
f.write("10101\n")
f.write("10101\n")
f.write("10101\n")
f.write("10021\n")
f.write("5\n")
f.write("10031\n")
f.write("10111\n")
f.write("10101\n")
f.write("10101\n")
f.write("12001\n")
f.write("5\n")
f.write("00013\n")
f.write("01110\n")
f.write("21000\n")
f.write("01111\n")
f.write("00000\n")
f.close()
````

````python
# for Jupyter Notebook
###
f = open("input.txt", "r")
input = f.readline
###

directions = {
    'up': lambda x: [x[0]+1, x[1]],
    'down': lambda x: [x[0]-1, x[1]],
    'left': lambda x: [x[0], x[1]-1],
    'right': lambda x: [x[0], x[1]+1],
}

T = int(input())

for test_case in range(1, T + 1):
    
    # Stack 을 이용한 Depth First Search (DFS)
    dfs_stack = []
    goal = None
    
    # make maze
    maze_size = int(input())
    maze = [ [False for _ in range(maze_size)] for _ in range(maze_size)]
    for i in range(maze_size):
        row = input().strip("\n")
        for j, v in enumerate(row):
            if v == '0':
                maze[i][j] = True
            elif v == '2':
                dfs_stack.append([i,j])
            elif v == '3':
                goal = [i,j]

    result = 0
    while len(dfs_stack)!=0:
        
        a = dfs_stack.pop()
        
        # find a's adjacent points
        for d in directions:
            ap = directions[d](a)
            
            if ap == goal:
                result = 1
                break
                
            if maze_size in ap or -1 in ap:
                continue
            
            else:
                if maze[ap[0]][ap[1]]:
                    maze[ap[0]][ap[1]] = False
                    dfs_stack.append(ap)
                    
    print("#"+str(test_case), result)
````

## 4880. [파이썬 S/W 문제해결 기본] 5일차 - 토너먼트 카드게임

※ SW Expert 아카데미의 문제를 무단 복제하는 것을 금지합니다.


사다리 게임이 지겨워진 알고리즘 반 학생들이 새로운 게임을 만들었다. 가위바위보가 그려진 카드를 이용해 토너먼트로 한 명을 뽑는 것이다. 게임 룰은 다음과 같다.
 

1번부터 N번까지 N명의 학생이 N장의 카드를 나눠 갖는다. 전체를 두 개의 그룹으로 나누고, 그룹의 승자끼리 카드를 비교해서 이긴 사람이 최종 승자가 된다.

그룹의 승자는 그룹 내부를 다시 두 그룹으로 나눠 뽑는데, i번부터 j번까지 속한 그룹은 파이썬 연산으로 다음처럼 두개로 나눈다.

(그림 생략)

두 그룹이 각각 1명이 되면 양 쪽의 카드를 비교해 승자를 가리고, 다시 더 큰 그룹의 승자를 뽑는 방식이다.

다음은 4명이 카드를 비교하는 경우로, 숫자 1은 가위, 2는 바위, 3은 보를 나타낸다. 만약 같은 카드인 경우 편의상 번호가 작은 쪽을 승자로 하고, 처음 선택한 카드는 바꾸지 않는다.

(그림 생략)

N명이 학생들이 카드를 골랐을 때 1등을 찾는 프로그램을 만드시오.


 

[입력]
 

첫 줄에 테스트 케이스 개수 T가 주어진다.  1≤T≤50
 

다음 줄부터 테스트 케이스의 별로 인원수 N과 다음 줄에 N명이 고른 카드가 번호순으로 주어진다. 4≤N≤100

카드의 숫자는 각각 1은 가위, 2는 바위, 3은 보를 나타낸다.

 

[출력]
 

각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, 1등의 번호를 출력한다.

```
입력:
3
4
1 3 2 1
6
2 1 1 2 3 3
7
1 3 3 3 1 1 3	 

출력:
#1 3
#2 5
#3 1
```

````python
# Make input file
f = open("input.txt", "w")
f.write("3\n")
f.write("4\n")
f.write("1 3 2 1\n")
f.write("6\n")
f.write("2 1 1 2 3 3\n")
f.write("7\n")
f.write("1 3 3 3 1 1 3\n")
f.close()
````

````python
# for Jupyter Notebook
###
f = open("input.txt", "r")
input = f.readline
###

T = int(input())
for test_case in range(1, T + 1):
    
    nn = [ 1+i for i in range(int(input()))]
    cards = list(map(int, input().split()))
        
    def fight(start, end):

        if start == end:
            return start

        leftwin = fight(start, (start + end)//2)
        rightwin = fight((start + end)//2+1, end)

        return rockPaperScissor(leftwin, rightwin)

    def rockPaperScissor(l, r):
        if cards[l-1] - cards[r-1] == 1:
            return l
        if cards[l-1] -cards[r-1] == -1:
            return r
        if cards[l-1] -cards[r-1] == 0:
            return l  
        if cards[l-1] -cards[r-1] == -2:
            return l
        if cards[l-1] -cards[r-1] == 2:
            return r
    
    print(f'#{test_case} {fight(nn[0], nn[-1])}')
````

※ SW Expert 아카데미의 문제를 무단 복제하는 것을 금지합니다.


NxN 배열에 숫자가 들어있다. 한 줄에서 하나씩 N개의 숫자를 골라 합이 최소가 되도록 하려고 한다. 단, 세로로 같은 줄에서 두 개 이상의 숫자를 고를 수 없다.

조건에 맞게 숫자를 골랐을 때의 최소 합을 출력하는 프로그램을 만드시오.
 

예를 들어 다음과 같이 배열이 주어진다.
 
| - | - | - |
|---|---|---|
| 2 | 1 | 2 |
| 5 | 8 | 5 |
| 7 | 2 | 2 |


이경우 1, 5, 2를 고르면 합이 8로 최소가 된다.

 

[입력]
 

첫 줄에 테스트 케이스 개수 T가 주어진다.  1≤T≤50
 

다음 줄부터 테스트 케이스의 첫 줄에 숫자 N이 주어지고, 이후 N개씩 N줄에 걸쳐 10보다 작은 자연수가 주어진다. 3≤N≤10

 

[출력]
 

각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, 합계를 출력한다.


````python
# Make input file
f = open("input.txt", "w")
f.write(
"""3
3
2 1 2
5 8 5
7 2 2
3
9 4 7
8 6 5
5 3 7
5
5 2 1 1 9
3 3 8 3 1
9 2 8 8 6
1 5 7 8 3
5 5 4 6 8   
""")
f.close()
````

````python
# for Jupyter Notebook
###
f = open("input.txt", "r")
input = f.readline
###

T = int(input())
for test_case in range(1, T + 1):
    
    array = []
    n = int(input())
    for _ in range(n):
        array.append( list( map(int, input().split())))
    
    sum_min = None
    
    def get_permutation(seq):
        
        global sum_min
        
        def create_permutation(seq, cur_seq, idx_used):
            global sum_min
            
            if sum_min != None:
                if sum_min < sum( [ array[i][cur_seq[i]] for i in range(len(cur_seq)) ]):
                    return
            if len(cur_seq) == len(seq):
                if sum_min == None:
                    sum_min = sum( [ array[i][cur_seq[i]] for i in range(n) ])
                else:
                    if sum_min > sum( [ array[i][cur_seq[i]] for i in range(n) ]):
                         sum_min = sum( [ array[i][cur_seq[i]] for i in range(n) ])
                return
            
            else:
                for i in range(len(seq)):
                    if not i in idx_used:
                        cur_seq.append(i)
                        idx_used.append(i)
                        create_permutation(seq, cur_seq, idx_used)
                        cur_seq.pop()
                        idx_used.remove(i)
        
        create_permutation(seq, [], [])
    
    get_permutation(range(n))
    
    print(f'#{test_case} {sum_min}')
    
````
